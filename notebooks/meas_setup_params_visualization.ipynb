# Visualizing measurement setup parameters

This notebook shows the meaning of the contents of a `simulator_setup.json` file (`SimulatorParams` object) on a schematic drawing.

In case of an existing measurement setup folder, you can use the `cam-bpw-sim draw-setup [PATH]` CLI command to get a similar drawing.

```python
import cam_bpw_sim as bps

bps.__version__
```

## Class 2 lever

For use with standard cams (the higher the waveform amplitude, the greater the cam radius).

```python
p = bps.meas.SimulatorParams(
    label="My_setup_1",
    lever_class=2,
    lever_r1=500,
    lever_r2=23,
    fulcrum_x=495,  # The eccentricity (epsilon) is added here
    cam_follower_d=1,
    cam_follower_h=0,
    lever_follower_d=8,
)
```

```python
p.draw(r_cam=30)
```

## Class 1 lever

This setup can be useful with cams with inverted signal.

```python
bps.meas.SimulatorParams(
    label="My_setup_2",
    lever_class=1,
    lever_r1=400,
    lever_r2=23,
    fulcrum_x=395,  # The eccentricity (epsilon) is added here
    cam_follower_d=1,
    cam_follower_h=0,
    lever_follower_d=8,
).draw(r_cam=30)
```
